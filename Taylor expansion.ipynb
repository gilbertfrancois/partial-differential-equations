## Taylor expansion

The Taylor expansion is defined as:

\begin{align}
f(x) &= f(a) + f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \frac{f'''(a)}{3!}(x-a)^3 + \cdots + \frac{f^{n}(a)}{n!}(x-a)^n + R_n(x) \\
f(x) &= \sum_{n=0}^{\infty} \frac{f^{n}(a)}{n!}(x-a)^n
\end{align}

### Example 1

Let's expand the function $f(x) = x^3$ around $x=1$:

\begin{align}
f(x) &= x^3 \\
f'(x) &= 3x^2 \\
f''(x) &= 6x \\
f'''(x) &= 6 \\
f^n(x) &= 0 \qquad \in n > 3
\end{align}

In this example, the taylor expansion gives an exact result, since the rest term $R_n(x) = 0$.

```python
def fn(x):
    return x**3

def dfn1(x):
    return 3*x**2

def dfn2(x):
    return 6*x

def dfn3(x):
    return 6
```

```python
def taylor(x, a):
    res = fn(a)
    res += 1/1 * dfn1(a)*(x-a)
    res += 1/2 * dfn2(a)*(x-a)**2
    res += 1/6 * dfn3(a)*(x-a)**3
    return res
```

```python
a = 1.0
x = np.linspace(-1, 3, 100)
```

```python
fig, axs = plt.subplots(1, 1, figsize=(14, 6))
axs.plot(x, fn(x), label="exact")
axs.plot(x, taylor(x, a), label="taylor expansion") 
axs.plot(a, fn(a), "go")
axs.legend()
axs.set_title("Taylor expansion around $x=1$")
plt.show()
```

### Example 2


\begin{align}
f(x) &= x^4 \\
f'(x) &= 4x^3 \\
f''(x) &= 12x^2 \\
f'''(x) &= 24x \\
\end{align}

Here, the rest term $R_n \neq 0$ with $n_{max} = 3$. The Taylor expansion error will increase with larger distance from its expansion point $a$.

```python
def fn(x):
    return x**4

def dfn1(x):
    return 4*x**3

def dfn2(x):
    return 12*x**2

def dfn3(x):
    return 24*x
```

```python
def taylor(x, a):
    res = fn(a)
    res += 1/1 * dfn1(a)*(x-a)
    res += 1/2 * dfn2(a)*(x-a)**2
    res += 1/6 * dfn3(a)*(x-a)**3
    return res
```

```python
a = 1.0
x = np.linspace(-1, 3, 100)
```

```python
fig, axs = plt.subplots(1, 1, figsize=(14, 6))
axs.plot(x, fn(x), label="exact")
axs.plot(x, taylor(x, a), label="taylor expansion")
axs.plot(a, fn(a), "go")
axs.legend()
axs.set_title("Taylor expansion around $x=1$")
plt.show()
```
